```python
class Vehiculo:
    def __init__(self, marca="", modelo="", anio=0):
        self.marca = marca
        self.modelo = modelo
        self.anio = anio

    def leer_datos(self):
        print(f"\n--- Ingresando datos del vehículo ---")
        self.marca = input("Introduce la marca: ")
        self.modelo = input("Introduce el modelo: ")
        self.anio = input("Introduce el año: ")

    def mostrar_datos(self):
        print(f"\n✅ Ficha Técnica del Vehículo:")
        print(f"🚗 Marca: {self.marca}")
        print(f"📂 Modelo: {self.modelo}")
        print(f"📅 Año: {self.anio}")
        print("------------------------------")

# --- Lógica de Interacción ---

# 1. Creamos los 2 objetos (instancias) solicitados
vehiculo1 = Vehiculo()
vehiculo2 = Vehiculo()

print("🏁 Registro de Vehículos en el Sistema")

# Capturando datos para el primer objeto
print("\nCONFIGURACIÓN VEHÍCULO 1")
vehiculo1.leer_datos()

# Capturando datos para el segundo objeto
print("\nCONFIGURACIÓN VEHÍCULO 2")
vehiculo2.leer_datos()

# Mostrando los resultados finales
print("\n" + "="*30)
print("REPORTE FINAL DE VEHÍCULOS")
print("="*30)

vehiculo1.mostrar_datos()
vehiculo2.mostrar_datos()
```
